```python
import pandas as pd

def wmape(actual, forecast):
    return (abs(actual - forecast)).sum() / actual.sum()

actual_values = pd.Series(df['Actual sales'])
forecast_values = pd.Series(df['Fct Value'])

wmape_score = wmape(actual_values, forecast_values)
print("WMAPE:", wmape_score)
```

Acuuracy of the forecast by using WMAPE is 61%

```python
def mae(actual, forecast):
    return (abs(actual - forecast)).mean()

actual_values = pd.Series(df['Actual sales'])
forecast_values = pd.Series(df['Fct Value'])

mae_score = mae(actual_values, forecast_values)
print("MAE:", mae_score)
```

```python
def smape(actual, forecast):
    return np.mean(200 * np.abs(forecast - actual) / (np.abs(forecast) + np.abs(actual)))

actual_values = pd.Series(df['Actual sales'])
forecast_values = pd.Series(df['Fct Value'])

smape_score = smape(actual_values, forecast_values)
print("SMAPE:", smape_score)

```

```python
def calculate_mape(actual_value, forecast_value):
    absolute_percentage_error = abs((actual_value - forecast_value) / actual_value)
    mape = absolute_percentage_error * 100
    return mape
```

```python
def mase(actual, forecast, naive_forecast):
    mae_actual = (abs(actual - forecast)).mean()
    mae_naive = (abs(actual - naive_forecast)).mean()
    return mae_actual / mae_naive
```

```python
actual=pd.Series([5,6])
forecast=pd.Series([10,12])
naive_forecast=pd.Series([7,8])
mase_error=mase(actual,forecast,naive_forecast)
```
